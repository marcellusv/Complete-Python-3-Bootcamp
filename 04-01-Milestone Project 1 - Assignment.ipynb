```python
from IPython.display import clear_output

def display_board(b):
    print('Tick Tac Toe board:')
    print(f'{b[0]} {b[1]} {b[2]}')
    print(f'{b[3]} {b[4]} {b[5]}')
    print(f'{b[6]} {b[7]} {b[8]}')

def get_board_position():
    position = 0
    while True:
        position = input('Enter a position on the board (1-9): ')
        # Reset position to 0 when player does not provide input (enter)
        if position == '':
            position = 0
        # Convert position to integer, because input() always returns a string
        position = int(position)
        # Validate position for board
        if position in range(1, 10):
            return position
        # Display message when position falls outside board
        print(f'Position {position} is outside range')

def update_board(game_board, position, second_player_turn):
    player_icon = 'X'
    if second_player_turn:
        player_icon = 'O'
    if position in game_board:
        game_board[position - 1] = player_icon

def check_winner(game_board):
    if check_winner_horizontal(game_board):
        return True
    elif check_winner_vertical(game_board):
        return True
    elif check_winner_diagonal(game_board):
        return True
    return False

def check_winner_horizontal(game_board):
    for row in range(0, 7, 3):
        if (
            game_board[row] == game_board[row + 1] and
            game_board[row] == game_board[row + 2]
        ):
            return True
    return False

def check_winner_vertical(game_board):
    for column in range(0, 3, 1):
        if (
            game_board[column] == game_board[column + 3] and
            game_board[column] == game_board[column + 6]
        ):
            return True
    return False

def check_winner_diagonal(game_board):
    if (
        (game_board[0] == game_board[4] and game_board[0] == game_board[8]) or
        (game_board[2] == game_board[4] and game_board[2] == game_board[6])
    ):
        return True
    return False

def check_tie(game_board):
    for i in range(1, len(game_board)+1, 1):
        if i in game_board:
            return False
    return True


# New game, initial board
game_board = [1,2,3,4,5,6,7,8,9]
# Variable to keep game running
game_on = True
second_player_turn = False

while game_on:
    # Clear game history and display the board
    clear_output()
    display_board(game_board)
    
    # Let player choose position
    position = get_board_position()

    # Update board
    update_board(game_board, position, second_player_turn)
    
    # Switch to next player
    if second_player_turn:
        second_player_turn = False
    else:
        second_player_turn = True

    # Player has won, end the game
    winner = check_winner(game_board)
    if winner:
        print('Winner!')
        game_on = False
    # Check for tie
    elif check_tie(game_board):
        print('Tie!')
        game_on = False
```

```python
def validate_horizontal_winner():
    test1 = check_winner_horizontal(['X','X','X',4,5,6,7,8,9])
    test2 = check_winner_horizontal([1,2,3,'X','X','X',7,8,9])
    test3 = check_winner_horizontal([1,2,3,4,5,6,'X','X','X'])
    print(test1, test2, test3)
def validate_vertical_winner():
    test1 = check_winner_vertical(['X',2,3,'X',5,6,'X',8,9])
    test2 = check_winner_vertical([1,'X',3,4,'X',6,7,'X',9])
    test3 = check_winner_vertical([1,2,'X',4,5,'X',7,8,'X'])
    print(test1, test2, test3)
def validate_diagonal_winner():
    test1 = check_winner_diagonal(['X',2,3,4,'X',6,7,8,'X'])
    test2 = check_winner_diagonal([1,2,'X',4,'X',6,'X',8,9])
    print(test1, test2)

validate_horizontal_winner()
validate_vertical_winner()
validate_diagonal_winner()
```
